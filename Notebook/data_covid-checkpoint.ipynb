```python
## Preprocessing
```

```python
#Dropping unnecessary columns
df2_covid = df_covid.loc[:,['date', 'granularite','maille_code','maille_nom','cas_confirmes','hospitalises','gueris','deces']]
#We will consider for this project cases by departement 
dfD_covid = df2_covid.loc[df2_covid['granularite']=="departement",:]
dfD_covid.head(50)

```

```python
##performing Data Cleaning
```

```python
dfD_covid.isna().sum()
```

```python
dfD_covid["date"] = pd.to_datetime(dfD_covid.date)
```

```python
dfD_covid.drop_duplicates()
```

```python
#et’s replace NAN with 0
dfD_covid['cas_confirmes']=dfD_covid['cas_confirmes'].fillna(0)
dfD_covid['deces']=dfD_covid['deces'].fillna(0)
dfD_covid['gueris']=dfD_covid['gueris'].fillna(0)
dfD_covid.isna().sum()
```

```python
grouped = []
for group in dfD_covid.groupby("maille_code"):
    grouped.append(group[1])
grouped
```

```python
full_grouped = dfD_covid.groupby(['date', 'maille_code'])['cas_confirmes', 'hospitalises', 'gueris', 'deces'].sum().reset_index()
full_grouped.head(n=20)
```

```python
full_grouped.to_csv('COVID-19_clean.csv')
```

```python
from sklearn.linear_model import LinearRegression
```

```python
dfD_covid.index = dfD_covid['date']
```

```python
x = dfD_covid["date"][:]
y = dfD_covid["cas_confirmes"][:]
x = x.values.reshape(-1, 1) #Eviter un problème de dimension : 2D array, 1D array 
```

```python
model = LinearRegression()
model.fit(x,y)
model.score(x,y)

prediction = model.predict(x)
```

```python
dfD= dfD_covid[['cas_confirmes']][:].copy()
dfD['prediction'] = prediction
dfD.head(5)
```

```python
dfD.plot(alpha=0.4, figsize=(15,10), color=['r','b'])
plt.legend(loc='upper left')
plt.show()
```
