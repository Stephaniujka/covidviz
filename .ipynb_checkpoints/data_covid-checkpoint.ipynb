```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from download import download 
```

```python
url1 ='https://www.data.gouv.fr/fr/datasets/r/0b66ca39-1623-4d9c-83ad-5434b7f9e2a4'
path_target = "./covid.csv"
download(url1, path_target, replace=True)
df_covid = pd.read_csv("covid.csv")
df_covid.head(n=100)
print(df_covid)
```

```python
#exctraction de colonnes du dataframe df_covid
df2_covid = df_covid.loc[:,['date', 'granularite','maille_code','maille_nom','cas_confirmes','hospitalises']]
#filtrage des données selon le departement
df2_covid = df2_covid.loc[df2_covid['granularite']=="departement",:]
```

```python
df2_covid.head(50)
```

```python
df2_covid["date"] = pd.to_datetime(df2_covid.date)
```

```python
df2_covid.dtypes
```

```python
index_with_nan = df2_covid.index[df2_covid.isnull().any(axis=1)]

print(index_with_nan)

df2_covid.drop(index_with_nan,0, inplace=True)

df2_covid.tail(100)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
df2_covid.index = df2_covid['date']
```

```python
x = df2_covid["date"][:]
y = df2_covid["cas_confirmes"][:]
x = x.values.reshape(-1, 1) #Eviter un problème de dimension : 2D array, 1D array 
```

```python
model = LinearRegression()
model.fit(x,y)
model.score(x,y)

prediction = model.predict(x)
```

```python
df2= df2_covid[['cas_confirmes']][:].copy()
df2['prediction'] = prediction
df2.head(5)
```

```python
df2.plot(alpha=0.4, figsize=(15,10), color=['r','b'])
plt.legend(loc='upper left')
plt.show()
```
